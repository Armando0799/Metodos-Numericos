*Definición:*
Sea $(\beta_n)$ una sucesión que converge a cero y sea $(a_n)$ tal que converge a $\alpha\in\mathbb{R}$. Si existe $k>0$ tal que
\begin{equation}
    |\alpha_n - \alpha|\leq k\ |\beta_n|
\end{equation}
para cada $n$, entonces decimos que $(a_n)$ converge a $\alpha$ con $\textbf{rapidez de convergencia}$ $\textbf{O}(\beta_n)$. Esta expresión se lee "O mayúscula de $\beta_n$" se indica escribiendo $\alpha_n=\alpha+O(\beta_n)$.

$\textbf{Ejemplo:}$ Sea $\displaystyle\alpha_n =\frac{n+1}{n^2}$ y $\displaystyle\tilde{\alpha}_n = \frac{n+3}{n^3}$. Notemos que ambas convergen a cero, sin embargo una sucesión converge más rápido a cero, para ver esto consideremos las sucesiones $\beta_n=\displaystyle\frac{1}{n}$ y $\tilde{\beta}_n=\displaystyle\frac{1}{n^2}$. Luego, $|\alpha_n-0|\leq 2 \beta_n$ y $|\tilde{\alpha}_n-0|\leq 2 \tilde{\beta}_n$ de modo que
\begin{equation}
    \alpha_n=0+O(\beta_n) \ \ \ \ y \ \ \ \tilde{\alpha}_n=0+O(\tilde{\beta}_n)
\end{equation}
Por tanto, $\tilde{\alpha}_n$ converge más rápido a cero.

$\textbf{Definición}:$
Supongamos que $\lim_{h\to 0} G(h)=0$ y $\lim_{h\to 0} F(h)=0$, si existe $k>0$ tal que $|F(h)-L|\leq k |G(h)|$ para $h$ sufiencientemente pequeña, entonces escribimos $F(h)=L+O(G(h))$.

$\textbf{Ejemplo}:$ Usando Taylor para $f(x)=\cos x$ y $n=3$ y $x_0$ se tiene:
\begin{equation}
    \cos(x)=1-\frac{1}{2}x^2+\frac{1}{24}x^4\cos(t_3)
\end{equation}
donde $t\in[0, x]$. Luego, notemos que
\begin{equation}
    \cos(x)+\frac{1}{2}x^2=1+O(x^4)
\end{equation}

\begin{equation}
    |\cos(x)+\frac{1}{2}x^2-1|=|\frac{1}{24}x^4\cos(t_3)|\leq\frac{1}{24}x^4
\end{equation}

Por lo tanto, cuando $x\longrightarrow 0$ se tiene que $\cos(x)-\frac{1}{2}\longrightarrow 1$ casi tan rápido como $x^4$ converge a cero.
