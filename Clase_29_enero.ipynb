
*  Error inherente: Datos maltomados.
*  Error de redondeo: Porque las computadoras pueden tomar un número finito de números.
*   Error por truncamiento: Por un número finito de instrucciones del algoritmo (La serie de Taylor).





**Definición:**
Sea $E_0$ un error inicial y $E_n$ el error después de $n$ operaciones sucesivas. Si $E_n=C E_0$, donde $C$ es una constante entonces el crecimiento del error es lineal. Si $E_n=C^n E_0$, donde $C>1$ entonces el crecimiento del error se denomina exponencial.

*Notas: Al algoritmo que muestra un crecimiento de error lineal se denomina estable e instable en caso contrario.*

Ejemplo: La ecuación recursiva $p_m=\frac{10}{3} p_{m-1}-p_{m-2}$, para $m=2, 3, \ldots$ la solución es:
  \begin{equation}
    p_m=c_1(\frac{1}{3})^m+c_2(3)^m
  \end{equation}

Si $p_0$=1 y $p_1=1/3$ entonces $c_1=1$ y $c_2=0$, es decir, $p_m=(1/3)^m$. Supongamos que usamos el redondeo por cuatro cifras entonces $\hat{p}_0=1.0000, \hat{p}_1=0.3333$ lo cual implica que $c_1=1.0000$ y $c_2=-0.12500\times10^{-5}$. Así,
\begin{equation}
  \hat{p}_m=1.0000(\frac{1}{3})^m-0.12500\times10^{-5}\ 3^m.
\end{equation}
y el error por redondeo es
\begin{equation}
  p_m-\hat{p}_m=0.12500\times10^{-5}\ 3^m.
\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np

m=np.linspace(0, 15, 100)

error=0.12500e-5*3**m

plt.figure()
plt.plot(m, error)
plt.xlabel("m")
plt.ylabel(r"$p_m-\hat{p}_m$")
plt.title("Gráfica de error por redondeo")
plt.grid()
plt.show()
```
